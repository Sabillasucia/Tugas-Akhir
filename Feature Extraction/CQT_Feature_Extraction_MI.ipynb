```python
import librosa
import librosa.display
import IPython.display as ipd
import numpy as np
from google.colab import drive
import os
import pandas as pd
import csv
import os
from glob import glob
import scipy.io.wavfile
import matplotlib.pyplot as plt
import pywt
import shutil
from pywt import wavedec
```

```python
drive.mount('/content/drive')
path_base = "/content/drive/MyDrive/Data Tugas Akhir"
Preprocess_MI = f'{path_base}{os.sep}Preprocess_MI/'
audio_files_MI = glob(Preprocess_MI  + '*.wav')
```

```python
def feature_extraction(file_path):
    x,sample_rate = librosa.load(file_path, res_type='kaiser_best')
    cqt = np.abs(librosa.cqt(x, sr=sample_rate))
    return cqt
```

```python
features_cqt = []
directory = Preprocess_MI
for audio in os.listdir(directory):
    audio_path = directory+audio
    features_cqt.append(feature_extraction(audio_path))
```

```python
features_cqt
```

Menghitung Mean

```python
cqt_means = []
for i in features_cqt:
  mean = np.mean(i)
  cqt_means.append(mean)
```

```python
df_mean = pd.DataFrame(cqt_means, columns=['Mean']) 
df_mean
```

```python
cqt_std = []
for i in features_cqt:
  std = np.std(i)
  cqt_std.append(std)
```

```python
df_std = pd.DataFrame(cqt_std, columns=['Std']) 
df_std
```

```python
cqt_max = []
for i in features_cqt:
  max = np.max(i)
  cqt_max.append(max)
```

```python
df_max = pd.DataFrame(cqt_max, columns=['Max']) 
df_max
```

```python
cqt_Min = []
for i in features_cqt:
  min = np.min(i)
  cqt_Min.append(min)
```

```python
df_min = pd.DataFrame(cqt_Min, columns=['Min']) 
df_min
```

```python
cqt_Med = []
for i in features_cqt:
  Med = np.median(i)
  cqt_Med.append(Med)
```

```python
df_med = pd.DataFrame(cqt_Med, columns=['Med']) 
df_mean
```

```python
cqt_Var = []
for i in features_cqt:
  Var = np.var(i)
  cqt_Var.append(Var)
```

```python
df_var = pd.DataFrame(cqt_Var, columns=['Var']) 
df_var
```

```python
from scipy.stats import skew
```

```python
cqt_Skew = []
for i in features_cqt:
  Skew = skew(i, axis=0, bias=True)
  cqt_Skew.append(Skew)
```

```python
skew_means = []
for i in cqt_Skew:
  mean = np.mean(i)
  skew_means.append(mean)
```

```python
df_skew = pd.DataFrame(skew_means, columns=['Skew']) 
df_skew
```

```python
cqt_Q1 = []
for i in features_cqt:
  Q1 = np.percentile(i, 25) 
  cqt_Q1.append(Q1)
```

```python
df_q1 = pd.DataFrame(cqt_Q1, columns=['Q1']) 
df_q1
```

```python
cqt_Q3 = []
for i in features_cqt:
  Q3 = np.percentile(i, 75) 
  cqt_Q3.append(Q3)
```

```python
df_q3 = pd.DataFrame(cqt_Q3, columns=['Q3']) 
df_q3
```

```python
cqt_IQR = []
IQR = np.array(cqt_Q3) - np.array(cqt_Q1)
cqt_IQR.append(IQR)
```

```python
df_iqr = pd.DataFrame(cqt_IQR)
df_iqr = df_iqr.transpose()
```

```python
cqt_range = []
Range = np.array(cqt_max) - np.array(cqt_Min)
cqt_range.append(Range)
```

```python
df_range = pd.DataFrame(cqt_range)
df_range = df_range.transpose()
```

```python
from scipy.stats import kurtosis
```

```python
cqt_Kurt = []
for i in features_cqt:
  Kurt = kurtosis(i, axis=0, bias=True)
  cqt_Kurt.append(Kurt)
```

```python
kurt_means = []
for i in cqt_Kurt:
  mean = np.mean(i)
  kurt_means.append(mean)
```

```python
df_kurt = pd.DataFrame(kurt_means, columns=['Kurt']) 
df_kurt
```

```python
df_cqt = pd.DataFrame()
df_cqt['Mean_cqt'] = df_mean['Mean']
df_cqt['Std_cqt'] = df_std['Std']
df_cqt['Max_cqt'] = df_max['Max']
df_cqt['Min_cqt'] = df_min['Min']
df_cqt['Med_cqt'] = df_med['Med']
df_cqt['Var_cqt'] = df_var['Var']
df_cqt['Skew_cqt'] = df_skew['Skew']
df_cqt['Q1_cqt'] = df_q1['Q1']
df_cqt['Q3_cqt'] = df_q3['Q3']
df_cqt['IQR_cqt'] = df_iqr[0]
df_cqt['MinMax_cqt'] = df_range[0]
df_cqt['Kurt_cqt'] = df_kurt['Kurt']
df_cqt
```
