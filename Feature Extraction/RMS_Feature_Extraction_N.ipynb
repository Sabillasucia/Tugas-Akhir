```python
import librosa
import librosa.display
import IPython.display as ipd
import numpy as np
from google.colab import drive
import os
import pandas as pd
import pywt
import csv
import os
from glob import glob
import shutil
from scipy.io import wavfile
import matplotlib
import matplotlib.pyplot as plt
```

```python
drive.mount('/content/drive')
directory = "/content"
base_path = "/content/drive/MyDrive/Data Tugas Akhir"
N_path = f'{base_path}{os.sep}N/'
audio_files_N = glob(N_path  + '*.wav')
Preprocess_N = f'{base_path}{os.sep}Preprocess_N/'
audio_files_N = glob(Preprocess_N  + '*.wav')
```

```python
def RMS_feature(file_path):
    x,sample_rate = librosa.load(file_path, res_type='kaiser_best')
    RMS = librosa.feature.rms(y=x)
    return RMS
```

```python
featuresRMS = {}
features_RMS_test = []
directory2 = Preprocess_N
for audio in os.listdir(directory2):
    audio_path = directory2+audio
    featuresRMS[audio_path] = RMS_feature(audio_path)
    features_RMS_test.append(RMS_feature(audio_path))
```

```python
df_RMS_N = pd.DataFrame({'RMS' : features_RMS_test})
df_RMS_N.head(10)
```

```python
RMS_means = []
for i in features_RMS_test:
  mean = np.mean(i,dtype=np.float64)
  RMS_means.append(mean)

df_RMS_means = pd.DataFrame({'Means' : RMS_means})
df_RMS_means
```

```python
RMS_std = []
for i in features_RMS_test:
  std = np.std(i)
  RMS_std.append(std)

df_RMS_std = pd.DataFrame({'Std' : RMS_std})
df_RMS_std
```

```python
RMS_max = []
for i in features_RMS_test:
  max = np.max(i)
  RMS_max.append(max)

df_RMS_max = pd.DataFrame({'Max' : RMS_max})
df_RMS_max
```

```python
RMS_min = []
for i in features_RMS_test:
  Min = np.min(i)
  RMS_min.append(Min)

df_RMS_min = pd.DataFrame({'Min' : RMS_min})
df_RMS_min
```

```python
RMS_Med = []
for i in features_RMS_test:
  Med = np.median(i)
  RMS_Med.append(Med)

df_med_rms = pd.DataFrame(RMS_Med, columns=['Med']) 
df_med_rms
```

```python
RMS_Var = []
for i in features_RMS_test:
  Var = np.var(i)
  RMS_Var.append(Var)

df_var_rms = pd.DataFrame(RMS_Var, columns=['Var']) 
```

```python
from scipy.stats import skew
RMS_Skew = []
for i in features_RMS_test:
  Skew = skew(i, axis=0, bias=True)
  RMS_Skew.append(Skew)
```

```python
skew_means_RMS = []
for i in RMS_Skew:
  mean = np.mean(i)
  skew_means_RMS.append(mean)
```

```python
df_skew_rms = pd.DataFrame(skew_means_RMS, columns=['Skew']) 
```

```python
RMS_Q1 = []
for i in features_RMS_test:
  Q1 = np.percentile(i, 25) 
  RMS_Q1.append(Q1)
```

```python
df_q1_rms = pd.DataFrame(RMS_Q1, columns=['Q1']) 
```

```python
RMS_Q3 = []
for i in features_RMS_test:
  Q3 = np.percentile(i, 75) 
  RMS_Q3.append(Q3)
```

```python
df_q3_rms = pd.DataFrame(RMS_Q3, columns=['Q3']) 
```

```python
RMS_IQR = []
IQR = np.array(RMS_Q3) - np.array(RMS_Q1)
RMS_IQR.append(IQR)

df_iqr = pd.DataFrame({'IQR' : RMS_IQR})
df_iqr = df_iqr.transpose()
df_iqr
```

```python
RMS_range = []
Range = np.array(RMS_max) - np.array(RMS_min)
RMS_range.append(Range)
```

```python
df_range = pd.DataFrame(RMS_range)
df_range = df_range.transpose()
df_range
```

```python
from scipy.stats import kurtosis
```

```python
RMS_Kurt = []
for i in features_RMS_test:
  Kurt = kurtosis(i, axis=0, bias=True)
  RMS_Kurt.append(Kurt)
```

```python
kurt_means = []
for i in RMS_Kurt:
  mean = np.mean(i)
  kurt_means.append(mean)
```

```python
df_kurt = pd.DataFrame({'Kurt' : kurt_means})
df_kurt
```

```python
df_rms = pd.DataFrame()
df_rms['Mean_rms'] = df_RMS_means['Means']
df_rms['Std_rms'] = df_RMS_std['Std']
df_rms['Max_rms'] = df_RMS_max['Max']
df_rms['Min_rms'] = df_RMS_min['Min']
df_rms['Med_rms'] = df_med_rms['Med']
df_rms['Var_rms'] = df_var_rms['Var']
df_rms['Skew_rms'] = df_skew_rms['Skew']
df_rms['Q1_rms'] = df_q1_rms['Q1']
df_rms['Q3_rms'] = df_q3_rms['Q3']
df_rms['IQR_rms'] = df_iqr[0]
df_rms['MinMax_rms'] = df_range[0]
df_rms['Kurt_rms'] = df_kurt['Kurt']
df_rms
```

```python
df_rms.to_csv("RMS_Feature Extraction_N.csv")
```
