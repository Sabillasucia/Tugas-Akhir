```python
import librosa
import librosa.display
import IPython.display as ipd
import numpy as np
from google.colab import drive
import os
import pandas as pd
import csv
import os
from glob import glob
import scipy.io.wavfile
import matplotlib.pyplot as plt
import pywt
import shutil
from pywt import wavedec
```

```python
drive.mount('/content/drive')
path_base = "/content/drive/MyDrive/Data Tugas Akhir"
Preprocess_MI = f'{path_base}{os.sep}Preprocess_MI/'
audio_files_MI = glob(Preprocess_MI  + '*.wav')
```

```python
def feature_extraction(file_path):
    x,sample_rate = librosa.load(file_path, res_type='kaiser_best')
    cr = librosa.feature.chroma_stft(y=x, sr=sample_rate)
    return cr
```

```python
features_cr = []
directory = Preprocess_MI
for audio in os.listdir(directory):
    audio_path = directory+audio
    features_cr.append(feature_extraction(audio_path))
```

```python
features_cr
```

Menghitung Mean

```python
cr_means = []
for i in features_cr:
  mean = np.mean(i)
  cr_means.append(mean)
```

```python
df_mean = pd.DataFrame(cr_means, columns=['Mean']) 
df_mean
```

```python
cr_std = []
for i in features_cr:
  std = np.std(i)
  cr_std.append(std)
```

```python
df_std = pd.DataFrame(cr_std, columns=['Std']) 
df_std
```

```python
cr_max = []
for i in features_cr:
  max = np.max(i)
  cr_max.append(max)
```

```python
df_max = pd.DataFrame(cr_max, columns=['Max']) 
df_max
```

```python
cr_Min = []
for i in features_cr:
  min = np.min(i)
  cr_Min.append(min)
```

```python
df_min = pd.DataFrame(cr_Min, columns=['Min']) 
df_min
```

```python
cr_Med = []
for i in features_cr:
  Med = np.median(i)
  cr_Med.append(Med)
```

```python
df_med = pd.DataFrame(cr_Med, columns=['Med']) 
df_mean
```

```python
cr_Var = []
for i in features_cr:
  Var = np.var(i)
  cr_Var.append(Var)
```

```python
df_var = pd.DataFrame(cr_Var, columns=['Var']) 
df_var
```

```python
from scipy.stats import skew
```

```python
cr_Skew = []
for i in features_cr:
  Skew = skew(i, axis=0, bias=True)
  cr_Skew.append(Skew)
```

```python
skew_means = []
for i in cr_Skew:
  mean = np.mean(i)
  skew_means.append(mean)
```

```python
df_skew = pd.DataFrame(skew_means, columns=['Skew']) 
df_skew
```

```python
cr_Q1 = []
for i in features_cr:
  Q1 = np.percentile(i, 25) 
  cr_Q1.append(Q1)
```

```python
df_q1 = pd.DataFrame(cr_Q1, columns=['Q1']) 
df_q1
```

```python
cr_Q3 = []
for i in features_cr:
  Q3 = np.percentile(i, 75) 
  cr_Q3.append(Q3)
```

```python
df_q3 = pd.DataFrame(cr_Q3, columns=['Q3']) 
df_q3
```

```python
cr_IQR = []
IQR = np.array(cr_Q3) - np.array(cr_Q1)
cr_IQR.append(IQR)
```

```python
df_iqr = pd.DataFrame(cr_IQR)
df_iqr = df_iqr.transpose()
```

```python
cr_range = []
Range = np.array(cr_max) - np.array(cr_Min)
cr_range.append(Range)
```

```python
df_range = pd.DataFrame(cr_range)
df_range = df_range.transpose()
```

```python
from scipy.stats import kurtosis
```

```python
cr_Kurt = []
for i in features_cr:
  Kurt = kurtosis(i, axis=0, bias=True)
  cr_Kurt.append(Kurt)
```

```python
kurt_means = []
for i in cr_Kurt:
  mean = np.mean(i)
  kurt_means.append(mean)
```

```python
df_kurt = pd.DataFrame(kurt_means, columns=['Kurt']) 
df_kurt
```

```python
df_cr = pd.DataFrame()
df_cr['Mean_cr'] = df_mean['Mean']
df_cr['Std_cr'] = df_std['Std']
df_cr['Max_cr'] = df_max['Max']
df_cr['Min_cr'] = df_min['Min']
df_cr['Med_cr'] = df_med['Med']
df_cr['Var_cr'] = df_var['Var']
df_cr['Skew_cr'] = df_skew['Skew']
df_cr['Q1_cr'] = df_q1['Q1']
df_cr['Q3_cr'] = df_q3['Q3']
df_cr['IQR_cr'] = df_iqr[0]
df_cr['MinMax_cr'] = df_range[0]
df_cr['Kurt_cr'] = df_kurt['Kurt']
df_cr
```

```python
df_cr.to_csv("cr_Feature Extraction_MI.csv")
```
